## Data Cleaner: <a name="datacleaner"></a>

```python
def clear_data(df):
    """
    Cria uma função que efetuada o data cleaner com os passos até 
    aqui aplicados
    """
    # Transformar em valores númericos
    df["VALOR_REEMBOLSADO"] = df["VALOR_REEMBOLSADO"].str.replace(",",".")
    df['VALOR_REEMBOLSADO'] = pd.to_numeric(df['VALOR_REEMBOLSADO'])
    
    # Drop data pois tenho mês e anod
    df = df.drop("DATA", axis=1)
    
    # Formatação dos dados em Tipo de despesas
    format_values = {'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.': 'Aluguel',
 'Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ': 'Despesas gerais',
 'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar': 'Contratações',
 'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes': 'Locomoção, hospedagem e alimentação',
 'Passagens aéreas, aquáticas e terrestres nacionais': 'Passagens',
 'Divulgação da atividade parlamentar': 'Divulgação',
 'Serviços de Segurança Privada': 'Segurança'}
    
    df['TIPO_DESPESA'] = df['TIPO_DESPESA'].map(format_values)
    df = df[["ANO", "MES", "SENADOR", "TIPO_DESPESA", "VALOR_REEMBOLSADO"]]
    
    english_names = ["YEAR", "MONTH", "SENATOR", "EXPENSE_TYPE", "REFUNDED_VALUE"]
    df.columns = english_names
    
    return df
```

Map the senator's name with your party.

```python
def partido_senador(df, siglas):
    """
    INPUT:
    df é um dataframe com os gastos dos senadores
    siglas é outro dataframe com mapeamento dos nomes dos senadores e seus partidos
    
    OUTPUT:
    retorna um dataframe com os gastos e os partidos dos senadores
    """
    partido = []
    for x in df["SENATOR"]:
        partido.append(siglas[siglas["Name"]==x]["Partido"].item())
        
    df["PARTY"] = partido
    
    return df
```

```python
df_2018 = partido_senador(df_2018, siglas)
```

## Exploratory Analysis:

**What is the total spending in 2018?**

```python
# Valor total de despesas com reembolso no ano atul de analise
total_2018 = df_2018["REFUNDED_VALUE"].sum()

print(f"The total spending in 2018 was {total_2018}")
```

**What is the change in total spending in recent years?**

```python
def gasto_total(list_df):
    """
    INPUT:
        Lista com varios dataframes.
        [df1, df2]
    
    OUTPUT:
        Dataframe com a soma total de reembolso por ano.
    """
    df = pd.DataFrame()
    for x in list_df:
        
        x = clear_data(x) # Chama a função para limpar o dado.
        
        x = x.pivot_table(index="YEAR", values="REFUNDED_VALUE", aggfunc=np.sum)
        df = df.append(x)
    return df
```

```python
gasto_total_ano = gasto_total([df_2014, df_2015,df_2016, df_2017])
```

```python
# Adiciona o ano em analise aos gasto_toal_ano
gasto_total_ano = gasto_total_ano.append(df_2018.pivot_table(index="YEAR", values="REFUNDED_VALUE", aggfunc=np.sum))
```

```python
# Plot gastos totais por ano
ax = plt.figure().gca()
ax.plot(gasto_total_ano, linewidth = 10)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.title("What is the change in total spending in recent years?", fontsize=30, y=1)
plt.xlabel("YEAR", fontsize=20)
plt.ylabel("Refunded Value", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.ticklabel_format(axis="both", style="plain")

sns.despine()
```

Spending has been increasing since 2014. We had a decrease in 2018. However 2018 is still a year of spending greater than 2016, 2015 and 2014.

### Analysis of 2018 spending.

```python
df_2018.groupby("EXPENSE_TYPE")["REFUNDED_VALUE"].sum()
```

```python
# Quais os tipos de despesas com maiores gastos
df_2018.groupby("EXPENSE_TYPE")["REFUNDED_VALUE"].sum().plot(kind="bar", fontsize=18)
plt.title("Expenses by expense type", fontsize=30)
plt.xlabel("EXPENSE_TYPE", fontsize=20)
plt.ylabel("REFUNDED VALUE", fontsize=20)
#plt.xticks(rotation=70)
sns.despine(bottom=True, left=True)
plt.savefig(img+'expenses_by_expense_type.png')
```

Brazilian senators spend the weekend in their home states and during the week work in Brasilia, the federal capital of Brazil. Therefore the high expense with passage.

Refunded values per month

```python
# Valores reembolsados em cada mês.
# Engraçado notar que o maior valor esta no mês 12 cujo senado está em
# recesso.
df_2018.pivot_table(index="MONTH", values="REFUNDED_VALUE", aggfunc=np.sum).plot(kind="bar", legend=False, fontsize=18)
plt.title("Spending per month", fontsize=30)
plt.xlabel("Month", fontsize=20)
plt.ylabel("Refunded value", fontsize=20)
sns.despine(bottom=True, left=True)
```

Month 12 is the highest expense month. But it's recess period!

**Which senators spent the most in 2018?**

```python
# Qual os 10 Senadores que mais gastaram em 2018?
MyPlot = df_2018.groupby("SENATOR")["REFUNDED_VALUE"].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", fontsize=21, figsize=(15,10))
plt.title("TOP 10", fontsize=30, pad=30)
plt.xlabel("Refunded Value", fontsize=25)
plt.ylabel("Senator", fontsize=25)
sns.despine(bottom=True, left=True)
```

## TOP 10 - By Expense Type

```python
fig, axarr = plt.subplots(4, 2, figsize= (15, 30))

MyPlot = df_2018[df_2018["EXPENSE_TYPE"]=="Daily Spend"].groupby('SENATOR')['REFUNDED_VALUE'].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("Daily Spend", fontsize=18)
axarr[0][0].set_ylabel("")
axarr[0][0].tick_params(axis ='x', rotation = 45) 


MyPlot = df_2018[df_2018["EXPENSE_TYPE"]=="Airline tickets"].groupby('SENATOR')['REFUNDED_VALUE'].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("Airline tickets", fontsize=18)
axarr[0][1].set_ylabel("")
axarr[0][1].tick_params(axis ='x', rotation = 45) 

MyPlot = df_2018[df_2018["EXPENSE_TYPE"]=="Rent"].groupby('SENATOR')['REFUNDED_VALUE'].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("Rent", fontsize=18)
axarr[1][0].set_ylabel("")
axarr[1][0].tick_params(axis ='x', rotation = 45) 

MyPlot = df_2018[df_2018["EXPENSE_TYPE"]=="General expenses"].groupby('SENATOR')['REFUNDED_VALUE'].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("General expenses", fontsize=18)
axarr[1][1].set_ylabel("")
axarr[1][1].tick_params(axis ='x', rotation = 45) 

MyPlot =df_2018[df_2018["EXPENSE_TYPE"]=="Hiring"].groupby('SENATOR')['REFUNDED_VALUE'].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("Hiring", fontsize=18)
axarr[2][0].set_ylabel("")
axarr[2][0].tick_params(axis ='x', rotation = 45) 

MyPlot = df_2018[df_2018["EXPENSE_TYPE"]=="Advertising"].groupby('SENATOR')['REFUNDED_VALUE'].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("Advertising", fontsize=18)
axarr[2][1].set_ylabel("")
axarr[2][1].tick_params(axis ='x', rotation = 45) 

MyPlot = df_2018[df_2018["EXPENSE_TYPE"]=="Security"].groupby('SENATOR')['REFUNDED_VALUE'].sum().nlargest(10)
MyPlot.sort_values(ascending=True).plot(kind="barh", ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("Security", fontsize=18)
axarr[3][0].set_ylabel("")
axarr[3][0].tick_params(axis ='x', rotation = 45) 


fig.delaxes(axarr[3,1])



plt.xticks(rotation=60)

fig.suptitle('TOP 10 - By Expense Type', y=.92,  fontsize=30)
plt.subplots_adjust(hspace=.4, wspace=.8, bottom=0.15)
sns.despine(bottom=True, left=True)
```

Expense with security is small. The Senator Fernando Collor was president of Brazil. However expenses with hiring catches the eye.
